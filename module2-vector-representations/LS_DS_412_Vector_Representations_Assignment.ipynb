## 2) Use Spacy to tokenize / clean the listings 

```python
nlp = spacy.load("en_core_web_lg")
tokenizer = Tokenizer(nlp.vocab)
```

```python
from bs4 import BeautifulSoup

def souper(df):
    soup = BeautifulSoup(df, 'html.parser')
    clean = soup.get_text(separator=" ")
    return clean

df['tokens'] = df.description.apply(souper)
df.head()
```

```python
tokens = []

""" Make them tokens """
for doc in tokenizer.pipe(df['tokens'], batch_size=500):
    doc_tokens = [token.text for token in doc]
    tokens.append(doc_tokens)

df['tokens'] = tokens
```

```python
df.tokens
```

## 3) Use Scikit-Learn's CountVectorizer to get word counts for each listing.

```python
text = df.tokens
```

```python
vector = CountVectorizer()

vector.fit(text)

dtm = vector.transform(text)
```

```python
print(vector.get_feature_names())
```

```python
dtm_df = pd.DataFrame(dtm.todense(), columns=vector.get_feature_names())
dtm_df
```

## 5) Use Scikit-Learn's tfidfVectorizer to get a TF-IDF feature matrix

```python
tfidf = TfidfVectorizer(stop_words='english')

dtm = tfidf.fit_transform(text)

dtm = pd.DataFrame(dtm.todense(), columns=tfidf.get_feature_names())
```

```python
dtm.head(10)
```

```python
def tokenize(document):
    
    doc = nlp(document)
    
    return [token.lemma_.strip() for token in doc if (token.is_stop != True) and (token.is_punct != True)]
```

```python
tfidf = TfidfVectorizer(stop_words='english', 
                        ngram_range=(1,2),
                        max_df=.97,
                        min_df=3,
                        tokenizer=tokenize)

dtm = tfidf.fit_transform(text)

dtm = pd.DataFrame(dtm.todense(), columns=tfidf.get_feature_names())
```
